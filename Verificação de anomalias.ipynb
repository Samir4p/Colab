```python
# 1- importando mas bibliotes do python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Importando as bibliotecas SKLEARN
from sklearn.neighbors import LocalOutlierFactor
#Importando um dataset da bibliotec SKLEARN
from sklearn.datasets import make_blobs
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
#gerando uma amostra de dados 200 linha e 2 fators apenas 
from scipy.sparse.construct import rand
dados, rotulos = make_blobs(n_samples=250, centers=1, n_features=2, random_state=0, center_box=(10,10), cluster_std=.3 )
```

```python
#Analisando as 10 primeiras linhas dos dados
dados[:10]
```

```python
#Plotamos os dados para averificação passando eixo y e x
plt.scatter(dados[:,0], dados[:,1])
plt.show()
```

```python
#Criando duas variaveis para gerar o modelo de predição
k_vizinhos = 20
contaminacao = .03
lof = LocalOutlierFactor(n_neighbors=k_vizinhos, contamination=contaminacao)
```

```python
#Vamos usar o algoritimo de prediçao
predicao = lof.fit_predict(dados)
```

```python
#Analisar os outliers da base de dados 
predicao
```

```python

lof_indice = np.where(predicao==-1)
anomalias = dados[lof_indice]
```

```python
fig,  eixo = plt.subplots()

eixo.scatter(dados[:,0], dados[:, 1])
eixo.scatter(anomalias[:,0], anomalias[:,1], color='red')
for i, texto in enumerate(anomalias):
  eixo.annotate('{:.2f}'.format(texto[0])+', '+'{:.2f}'.format(texto[1]), (anomalias[i, 0], anomalias[i,1]))

plt.show()  
```
